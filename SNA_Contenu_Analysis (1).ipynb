**Counting top words (post_text)**

```python
import string
```

```python
def count_words(df):

  mapping = dict.fromkeys(map(ord,string.punctuation))
  #key,value
  words={}
  for m in df['post_text']:
      try:
        separated_words=m.split(' ')
      except:
        pass
      for word in separated_words:
          word = word.strip()
          word = word.lower()
          word=word.translate(mapping) #emove any punctuation
          if len(word)<5:
            continue
          if word in words:
            words[word]+=1 
          else:
            words[word]=1
  print("done counting words")
  return words
```

```python
words=count_words(df)
```

```python
words
```

```python
words_df = pd.DataFrame(words.items(), columns=['word','count'])
words_df
```

```python
words_df_top = words_df.sort_values('count',ascending=False)
words_df_top
```

```python
%matplotlib inline
words_df[words_df['count']>10].plot(x='word', kind='bar',figsize=(15, 10))
```
